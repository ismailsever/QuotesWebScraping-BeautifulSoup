```python
#Import Libraries
from bs4 import BeautifulSoup as bs
import pandas as pd
pd.set_option('display.max_colwidth', 750)
import time
import requests
import random
```

# PAGE 1 - WEB SCRAPING

```python
#Accessing the website for page 1
page = requests.get('https://www.kdnuggets.com/2017/05/42-essential-quotes-data-science-thought-leaders.html')
page
```

```python
#Parsing the website
soup = bs(page.content,'html.parser')
print(soup.prettify())
```

```python
#Finding all instances of a tag at once/Exract the "qutoes"/Turn them into text
quotes1 = soup.find(id = 'post-')
quotes2 = quotes1.find_all('li')
quotes3 = [i.text for i in quotes2]
quotes3
```

```python
#Exract the "authors"/Turn them into text
authors = quotes1.select('a em')
authors1 = [i.text for i in authors]
authors1
```

```python
#Bring it all together and create the dataframe
q1 = pd.DataFrame({
    "Quotes": quotes3,
    "Authors": authors1
})
q1
```

# PAGE 2 - WEB SCRAPING

```python
#Accessing the website for page 2
page2 = requests.get('https://www.kdnuggets.com/2017/05/42-essential-quotes-data-science-thought-leaders.html/2')
page2
```

```python
#Parsing the website
soup1 = bs(page2.content,'html.parser')
print(soup1.prettify())
```

```python
#Finding all instances of a tag at once/Exract the "qutoes"/Turn them into text
quotes4 = soup1.find(id = 'post-')
quotes5 = quotes4.find_all('li')
quotes6 = [i.text for i in quotes5]
quotes6
```

```python
#Exract the "authors"/Turn them into text
authors2 = quotes4.select('a em')
authors3 = [i.text for i in authors2]
authors3
```

```python
#Create the dataframe for Quotes
q2 = pd.DataFrame({
    "Quotes": quotes6,
})
q2
```

```python
#Drop the unnecessary values
q2f = q2.drop(q2.index[[20, 21, 22, 23, 24, 25, 26]])
q2f
```

```python
#Create the dataframe for Authors
q3 = pd.DataFrame({
    "Authors": authors3
})
q3
```

```python
#Drop the 19th row
q3f = q3.drop([q3.index[19]])
q3f
```

```python
#Replace the value
q3f['Authors'] = q3f['Authors'].replace(['Foster Provost'],'Foster Provost, Tom Fawcett')
q3f
```

```python
# Reset the index
q4f = q3f.reset_index(drop=True)
q4f
```

```python
#Merge Quotes and Authors of Page 2
q4 = [q2f, q4f]
q5 = pd.concat(q4,axis=1, join='inner')
display(q5)
```

# Merging Page 1 and Page 2

```python
#Merge two dataframes from Page 1 and Page 2
merge = q1.append(q5)
merge
```
